```python
import pandas as pd
#1. load dataset
df = pd.read_csv("house_prices.csv")
df.head()
```

```python
# Explore  the dataset
print(df.info())
print(df.describe())
print("missing values: \n", df.isnull().sum())
```

```python
#3. Select features (x), and target(y)
X = df[["size_sqft", "bedrooms"]]
Y = df["price"]
```

```python
#4. Split into tarin and test data from skleaer
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size=0.2, random_state=42)
```

```python
print("training rows", len(X_train))
print("testing rows", len(X_test))
```

```python
#5.train the model
from sklearn.linear_model import LinearRegression
model = LinearRegression()
model.fit(X_train, y_train)
print("Model trained successfully")
```

```python
#6. Make predictions - Testing
y_pred = model.predict(X_test)
print("Predictions", y_pred)
print("Actual Values", y_test.values)
```

```python
#7. evaluate the model(mse, r2)
from sklearn.metrics import mean_squared_error, r2_score
mse = mean_squared_error(y_test, y_pred)
r2 = r2_score(y_test, y_pred)

print("Mean Squared Error", mse)
print("R-squared", r2)
```

```python
results = pd.DataFrame({
    "actual price": y_test.values,
    "Preedicted price": y_pred
})
results
```

```python
#8 Use model for prediction
new_house = pd.DataFrame([[1400,3]], columns=["size_sqft", "bedrooms"])
predicted_price = model.predict(new_house)
print("Predicted price for new_house:", predicted_price[0])
```
