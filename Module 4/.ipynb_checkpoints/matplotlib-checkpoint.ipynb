```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.DataFrame({
    "name": ["Asha", "Rahul", "Meera", "Arjun"],
    "age": [20, 19, 21, 20],
    "math_score": [85, 70, 92, 60]
})
print(df)
```

```python
# Line plot
plt.plot(df["age"], df["math_score"])
plt.xlabel("Age")
plt.ylabel("Math Score")
plt.title("Age vs Math Score")
plt.show()
```

```python
#Bar chart
plt.bar(df["name"], df['math_score'])
plt.xlabel("Name")
plt.ylabel("Math score")
plt.title("Math scores by students")
plt.show()
```

```python
#histogram
plt.hist(df["math_score"], bins=5)
plt.xlabel("Math score")
plt.ylabel("Count")
plt.title("Distribution of Math score")
plt.show()
```

```python
#Scatter plot
plt.scatter(df["age"], df["math_score"])
plt.xlabel("Age")
plt.ylabel("Math score")
plt.title("Scatter: Age Vs Math score")
plt.show()
```
