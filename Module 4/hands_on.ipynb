```python
# 1) Load dataset
import pandas as pd
df = pd.read_csv("student_scores.csv")
```

```python
# 2) Explore data
df.head()
df.info()
df.describe()
```

```python
# 3)Clean missing values
# Replace missing scores with 0

print(df)
df[["math_score","english_score","science_score"]] = df[["math_score","english_score","science_score"]].fillna(0)
```

```python
#4) Create total + average score columns
df["total"] = df["math_score"] + df["english_score"] + df["science_score"]
df["average"] = df["total"] / 3
```

```python
#5) NumPy calculations
import numpy as np

avg_scores = np.array(df["average"])
print("Class Average:", np.mean(avg_scores))
print("Highest Average:", np.max(avg_scores))
print("Lowest Average:", np.min(avg_scores))
```

```python
#6) Visualizations

#Matplotlib: Histogram of averages

import matplotlib.pyplot as plt

plt.hist(df["average"], bins=5)
plt.title("Distribution of Average Scores")
plt.xlabel("Average Score")
plt.ylabel("Count")
plt.show()
```

```python
#Seaborn: Scatter plot math vs science

import seaborn as sns

sns.scatterplot(data=df, x="math_score", y="science_score")
```

```python
#Seaborn: Correlation heatmap

sns.heatmap(df[["math_score","english_score","science_score","age","average"]].corr(), annot=True)
```

```python
#7) Save final cleaned dataset
df.to_csv("student_scores_cleaned.csv", index=False)
```
