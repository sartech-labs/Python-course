```python
import pandas as pd

df = pd.DataFrame({
    "name": ["Asha", "Rahul", "Meera", "Arjun"],
    "age": [20, 19, 21, 20],
    "math_score": [85, 70, 92, 60]
})
print(df)
```

```python
import seaborn as sns
```

```python
sns.scatterplot(data=df, x="age", y="math_score")
```

```python
sns.barplot(data=df, x="name", y="math_score")
```

```python
sns.histplot(df["math_score"], bins=5)
```

```python
sns.heatmap(df[["age", "math_score"]].corr(), annot=True)
```
